### Introdução:

O projeto de experimentos é um dos pilares da prática estatística, o objetivo é projetar um experimento a fim de confirmar ou rejeitar uma hipótese

Etapas:

```
Formular Hipótese -> Desenhar experimento -> Coletar Dados -> Inferência/Conclusões
```

### Testagem A/B

#### Tratamento
- Algo que um indivíduo é exposto. Ex: droga, capa do filme, preço

#### Grupo de tratamento
- Grupo de indivíduos exposto ao tratamento

#### Grupo de controle
- Grupo de indivíduos exposto a nenhum tratamento (ou padrão).
- O grupo de controle é usado para ter uma referência e conseguir identificar diferenças devido ao tratamento (ou ao acaso).

#### Randomização
- Selecionar aleatoriamente os indivíduos para cada grupo

#### Indivíduos
- Os itens que serão expostos ao tratamento. Ex: visitantes do site, pacientes

#### Estatística de teste
- Métrica usada para medir o efeito do tratamento. Comumente é usada uma métrica binária: clicar/não clicar, comprar/não comprar.

### Testes de Hipótese (ou Teste de Significância)
- Tem o propósito de ajudar a descobrir se uma chance aleatória poderia ser responsável por um efeito observado.
- É uma forma de proteger os pesquisadores de serem enganados pela possibilidade aleatória.
- Avalia se a possibilidade aleatória é uma explicação plausível para a diferença observada entre os grupos estudados.

#### Hipótese Nula
- Hipótese cuja possibilidade é provada
- A natureza da hipótese nula determina a estrutura do teste de hipótese

#### Hipótese Alternativa
- Contraponto da nula (o que se espera provar)
- Em conjunto com a Hipótese Nula, devem representar todas as possibilidades.


#### Teste Unilateral
- Teste de hipótese que conta as possibilidades em apenas uma direção

#### Teste Bilateral
- Teste de hipótese que conta as possibilidades em duas direções


### Reamostragem
- Amostrar repetitivamente os valores a partir dos dados observados, com o objetivo geral de avaliar a variabilidade aleatória em uma estatística

#### Bootstrap
- É usado para avaliar a confiabilidade de uma estimativa

#### Testes de permutação
- É usado para testar hipóteses, geralmente envolvendo dois ou mais grupos.
- Exemplo: A diferença identificada entre dois, ou mais, grupo está dentro da faixa que o acaso poderia produzir ou, por outro lado, é estatisticamente significante.
- Passos:
    - Combinar os resultados dos grupos (A, B, C, D,...). Essa é a representação lógica da hipótese nula da qual os tratamentos não diferem. A combinação é realizada pois presupoe que a diferença se materá entre os grupos após a combinação dos grupos.
    - Extraia aleatoriamente uma reamostra de tamanho correspondente a cada grupo (A, B, C, D,...).
    - Calcular a estatística/estimativa, ou seja, calcular a diferença (por exemplo, das médias) entre os grupos, e armazena o resultado
    - Repita os passos anteriores para produzir uma distribuição de permutações da estatística de teste
    - Constroi um histograma e calcula quantos porcento dos valores são diferentes da hipótese nula e compara com o nível de significância.
    
#### Testes de permutação exaustiva e Bootstrap
- Gerar todas as divisões possíveis entre os grupos após combinação deles
- Indicado para amostras pequenas
- Em testes de permustação bootstrap há reposição na divisão dos grupos

### Significância Estatística e Valores P

#### Significância
- É como os estatísticos medem se um experimento produz um resultado mais extremo do que o acaso proderia produzir.

#### Valor P
- Dado um modelo de acaso que representa a hipótese nula, o valor p é a probabilidade de obter resultados tão incomuns ou extremos quanto os resultados observados

#### Alfa
- Um limiar de probabilidade de "ser incomum" que os resultados do acaso podem ultrapassar, para resultados reais serem definidos como estatisticamente significantes

#### Erro tipo1
- Concluir erroneamente que um efeito é real (quando acontece pelo acaso)
- Os testes de significância são estruturados para minizar o Erro tipo 1

#### Erro tipo 2
- Concluir erroneamente que um efeito ocorreu pelo acaso (quando é real)

#### Ciência de Dados e Valor P
- Para os cientista de dados o valor P é mais utilizado para validar um modelo ou decidir se uma característica deve ser considerada ou não no modelo de ML.

### Testes t
- É um teste de hipótese que usa conceitos estatísticos para rejeitar ou não uma hipótese nula quando a estatística de teste (t) segue uma distribuição t de Student.
- Teste t pode ser conduzido para:
    - Comparar uma amostra com uma população
    - Comparar duas amostras pareadas
    - Comparar duas amostras independentes 
- Premissas
    - Independência de observações
    - Normalidade
    - Variações iguais
- Passos:
    - Formule a hipótese.
    - Defina um nível de significância
    - Colete a prepare os dados
    - Calcule a estatística t
    - Determine o valor P
    - Tome uma decisão

#### Estatística de teste
- Uma métrica para a diferença ou efeito de interesse

#### Estatística t
- Uma versão padronizada da estatística de teste
- Se a variável de interesse segue uma distribuição próxima de uma curva normal em ambas populações,
    - $T = \dfrac{(\bar x_{1} - \bar x_{2})-(\mu_{1}-\mu_{2})}{\sqrt{S_{p}^2}(\dfrac{1}{n_{1}}+\dfrac{1}{n_{2}})}$
    - Segue uma distribuição t Student com n1+ n2-2 graus de liberdade
    - $S_{p}^{2}$ é uma média entre as variâncias amostrais $S_{1}^{2}$ e $S_{2}^{2}$ ponderada pelos respectivos graus de liberdade $n_{1}-1$ e $n_{2}-1$
- Então temos $S_{p}^{2} =\dfrac{(n_{1}-1)S_{1}^{2}+(n_{2}-1)S_{2}^{2}}{n_{1}+n_{2}-2} $
- Identifica o $t^{*}$ através da tabela de distribuição de acordo com os graus de liberdade e significância
- Compara o $T$ calculado com o $t^{*}$ para rejeitar ou não a hipótese nula
- Um teste t reduz todos os dados em um único valor, chamado estatística t. Este valor único serve como uma medida de evidência contra a hipótese declarada (H0). Uma estatística t próxima de zero representa a menor evidência contra a hipótese. Um valor maior da estatística t representa forte evidência contra a hipótese.
- Um valor p é a probabilidade percentual da estatística t ter ocorrido por acaso. É representado como um decimal, por exemplo, um valor p de 0,05 representa uma probabilidade de 5% de ver uma estatística t pelo menos tão extrema quanto a calculada, assumindo que a hipótese nula era verdadeira.

#### Distribuição t
- Uma distribuição de referência, com a qual a estatística t observada pode ser comparada

### Exemplo de Teste T de uma amostra
- Em testes estatísticos, um teste t de uma amostra é usado quando queremos comparar uma média amostral com uma média populacional. Um teste t de uma amostra examina se a média de uma amostra é estatisticamente diferente de uma média populacional conhecida ou hipotética.

````python
# Import necessary libraries
import numpy as np
from scipy import stats

# Given student scores
student_scores = np.array([72, 89, 65, 73, 79, 84, 63, 76, 85, 75])


# Hypothesized population mean
mu = 70

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(student_scores, mu)
print("T statistic:", t_stat)
print("P-value:", p_value)

# Setting significance level
alpha = 0.05

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis; there is a significant difference between the sample mean and the hypothesized population mean.")
else:
    print("Fail to reject the null hypothesis; there is no significant difference between the sample mean and the hypothesized population mean.")
````

### Exemplo de Teste T de duas amostras

````python
# Import the necessary libraries:
import seaborn as sns
import numpy as np
from scipy import stats

# Load the Iris dataset:
iris = sns.load_dataset('iris')

# Filter the dataset for the two species we want to compare:
setosa = iris[iris['species'] == 'setosa']
versicolor = iris[iris['species'] == 'versicolor']

# Extract the petal lengths for each species:
setosa_petal_lengths = setosa['petal_length']
versicolor_petal_lengths = versicolor['petal_length']

# Perform the t-test:
t_stat, p_value = stats.ttest_ind(setosa_petal_lengths, versicolor_petal_lengths)

print("T statistic:", t_stat)
print("P-value:", p_value)

# Interpret the results:
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis; there is a significant difference between the petal lengths of Iris setosa and Iris versicolor.")
else:
    print("Fail to reject the null hypothesis; there is no significant difference between the petal lengths of Iris setosa and Iris versicolor.")
````
